```python
import numpy as np
import matplotlib.pyplot as plt
import math as mt
```

```python
x = np.arange(10)
y = (x-5)**2
plt.plot(x,y)
plt.show()
```

```python
# min that mminimizes y (cost of error fn)
x_ = 0
learning_rate = 0.1
error = 0;
error_list = []
for i in range(30):
    x_ = x_ - learning_rate*2*(x_-5)
    error = (x_-5)**2
    error_list.append(error)
    print(x_)    
    print(error)
    print(" ")
    plt.scatter(x_,error)
plt.show()
```
