```
from helper import *
import numpy as np 
import matplotlib.pyplot as plt
```

```
X,Y = load_dataset('xor')
```

```
print(X.shape, Y.shape)
```

```
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```
from keras.models import Sequential
from keras.layers import *
```

```
model = Sequential()
model.add(Dense(6,input_shape=(2,),activation='tanh'))
model.add(Dense(4,activation='tanh'))
model.add(Dense(2,activation='softmax'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.summary()
```

```
#Train
from keras.utils import to_categorical
Y_oht = to_categorical(Y)
print(Y_oht.shape)
```

```
model.fit(X,Y_oht,epochs=300)
```

```
loss, acc = model.evaluate(X,Y_oht) #returns loss and accuracy
print("Accuracy ", acc)
```

```
plot_decision_boundary(lambda x: np.argmax(model.predict(x),axis=1),X,Y)
```
