```python
import numpy as np
import matplotlib.pyplot as plt
import math as mt
import pandas as pd
```

```python
X= np.load('LinearX.npy')
Y= np.load('LinearY.npy')
```

```python
print(X.shape)
print(Y.shape)
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
theta = np.zeros((2,))
print(theta)
```

```python
def getYpred(theta,X):
    return theta[1]*X + theta[0];
def getGrads(X,Y,theta):
    grads = np.zeros((2,))
    Y_ = getYpred(theta,X);
    m = X.shape[0];
    grads[0] = 2*np.sum(Y_-Y)
    grads[1] = 2*np.sum((Y_-Y)*X)
    
    return grads/m

def error(X,Y,theta):
    Y_ = getYpred(theta,X);
    m = X.shape[0]
    error = np.sum((Y-Y_)**2)/m
    return error/m 
```

```python
getGrads(X,Y,theta)
```

```python
def gradDescent(X,Y,lr):
    errorList = []
    for i in range(100):
        grad = getGrads(X,Y,theta)
        theta[0] = theta[0] - lr*grad[0]
        theta[1] = theta[1] - lr*grad[1]
        e = error(X,Y,theta)
        errorList.append(e)
    return theta, errorList
```

```python
theta, errorList = gradDescent(X,Y,0.1)
```

```python
plt.plot(errorList)
plt.show()
```

```python
plt.scatter(X,Y)
yPred = getYpred(theta,X)
plt.plot(X,yPred,color='red')
plt.show()
```
