```
import os 
os.listdir("Train/Pikachu/")
```

```
from keras.preprocessing import image
import matplotlib.pyplot as plt
```

```
img = image.load_img("Train/Balbasaur/425.jpg", target_size=(100,100))
img = image.img_to_array(img)/255.0
# img[:,:,1:] = 0;
plt.imshow(img)
plt.show()
```

```
print(img.shape)
```

```
from keras.models import Sequential
from keras.layers import *

model = Sequential()
model.add(Conv2D(32,(3,3),activation='relu',input_shape=(100,100,3)))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(128,(3,3),activation='relu'))
model.add(MaxPooling2D(2,2))

model.add(GlobalAveragePooling2D())


model.add(Dense(64,input_shape=(2,),activation='relu'))
# model.add(Dense(15,input_shape=(2,),activation='relu'))
model.add(Dense(3,activation='softmax'))
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])

model.summary()
```

```
train_datagen = image.ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    rescale = 1.0/255,
    horizontal_flip=True)
val_datagen = image.ImageDataGenerator(
  rescale = 1.0/255
)
```

```
train_loader = train_datagen.flow_from_directory('Train',batch_size=32,class_mode='categorical',target_size=(100,100),shuffle='true')
val_loader =  val_datagen.flow_from_directory('Test',batch_size=32,class_mode='categorical',target_size=(100,100),shuffle='true')
```

```
train_loader.classes
```

```
for x,y in train_loader:
  print(x.shape,y.shape)
  for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(x[i])
  plt.show()    
  break
```

```
hist = model.fit_generator(train_loader,epochs=50,steps_per_epoch=10, shuffle=True,validation_data=val_loader,validation_steps=2)
```
